## Task 7 : Finding Impact of Description On Revenue

```python
!pip install eli5
import eli5
from sklearn.linear_model import LinearRegression
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
```

```python
vectorizer = TfidfVectorizer(sublinear_tf=True,analyzer='word',token_pattern=r'\w{1,}',
                           ngram_range=(1,2),min_df=5)
```

```python
overview_text = vectorizer.fit_transform(train_data['overview'].fillna(''))
```

```python
# Creating Linear Regression model
model = LinearRegression()
model.fit(overview_text,train_data['log_revenue'])
eli5.show_weights(model,vec=vectorizer,top=20,feature_filter=lambda x:x!='<Bias>')
```

 The words shown in green above impact positively to revenue,
 whereas the words shown in red impact it in a negative manner
